```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'--')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 =fig.add_axes([0.2,0.5,.2,.2])
```

```python
axes1.plot(x,y)
axes2.plot(y,x)
fig
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1,])
ax2 =fig.add_axes([0.2,0.5,.4,.4])
```

```python
ax1.plot(x,z)
ax2.plot(x,y)
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
fig
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2,figsize=(12,2))
```

```python
axes[0].plot(x,y,color = "blue",lw = 3,ls = '--')
axes[1].plot(x,z,color = "red", lw = 2,ls = '-')
fig
```
