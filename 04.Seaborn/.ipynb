```python
import numpy as np
import seaborn as sns
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,hue='smoker',estimator = np.min)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data = tips,palette = 'rainbow')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette = 'coolwarm',hue = 'smoker')
```

```python
sns.boxplot(data = tips,palette = 'rainbow',orient = 'h')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True,palette = 'Set1')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False,hue='sex',palette='Set1')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='rainbow')
sns.swarmplot(x='day',y='total_bill',data=tips,color='black',size=3)
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='strip',jitter=False)
```
