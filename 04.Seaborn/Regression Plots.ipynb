```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data = tips,palette = 'rainbow',markers=['o','v'],scatter_kws={'s':100},col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data = tips,palette = 'rainbow',markers=['o','v'],scatter_kws={'s':100},row='sex',col='time',aspect=0.6,size=6)
```
